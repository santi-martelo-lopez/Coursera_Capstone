```python
#lets load relevant libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#lets find the table
wiki_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
table_class = 'wikitable sortable jquery-tablesorter'
```

```python
#lets download the page
response = requests.get(wiki_url)
soup1 = BeautifulSoup(response.text,'html.parser')
#table_toronto = soup1.find("table",{"class":"wikitable sortable jquery-tablesorter"})
table_toronto = soup1.find("table",attrs={"class":"wikitable sortable"})
#print(soup1)
#print(table_toronto)
```

```python
# lets create a data frame
df1 = pd.read_html(str(table_toronto))
#print(df1)
##df1 = pd.DataFrame(df1)
table_rows = table_toronto.find_all('tr')
l = []
for ii,tr in enumerate(table_rows):
    if(ii!=0): # lets skip the first row that in blank
        td = tr.find_all('td')
        row = [tr.text[:-1] for tr in td] # with [:-1] we get rid off the "\n" character
        l.append(row)
df1 = pd.DataFrame(l, columns=['Postal Code','Borough','Neighbourhood'])
pd.set_option('display.max_columns',10)
pd.set_option('display.max_rows',5000)
pd.set_option('display.min_rows',1000)
pd.set_option('display.width',1000)
pd.set_option('display.max_colwidth',100)
pd.set_option('expand_frame_repr',100)
#print(df1)
df1
```

lets filter out the rows in the neighbouhood column with "not assigned" status

```python
# lets filter out the rows in the neighbouhood column with "not assigned" status
df_filtered = df1[df1['Neighbourhood']!='Not assigned'].reset_index(drop=True)
df_filtered
```
